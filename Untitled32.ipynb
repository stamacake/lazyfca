```python
def make_intent(example):
    global attributes
    return set([i+':'+str(k) for i, k in zip(attributes, example)])
```

Read data

```python
def read_data(idx, idxt):  
    pluses, minuses=[],[]
    X_test,y_test=[],[]
    for i in range(idx):
        ii=str(i+1)
        f = open("data/train" + ii + ".csv", "r")
        train = [a.strip().split(",") for a in f]
        tmp1 = [make_intent(a[:-1]) for a in train if a[-1] == "positive"]
        tmp2 = [make_intent(a[:-1]) for a in train if a[-1] == "negative"]
        f.close()

        pluses+=tmp1
        minuses+=tmp2   
    for i in range(idxt):
        ii=str(i+1)
        f = open("data/test" + ii + ".csv", "r")
        _unknown = [a.strip().split(",") for a in f]

        tmp1 = [make_intent(a[:-1]) for a in _unknown]
        tmp2 = [1 if a[-1] == "positive" else 0 for a in _unknown ]
        del tmp1[0]
        del tmp2[0]
        f.close()
        X_test+=tmp1
        y_test+=tmp2
        
    return pluses, minuses, X_test, y_test
```

Algorithm norm sum of power of features intersection with positive and negative.

Get value that is bigger.

```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
np.seterr(divide='ignore', invalid='ignore')

TP1 = 0
TN1 = 0
FP1 = 0
FN1 = 0
TPR1 = 0
TNR1 = 0
FPR1 = 0
NPV1 = 0
FDR1 = 0
acc1 = 0
prec1 = 0
rec1 = 0
def algorithm0(pluses, minuses, X, y):
    global TP1
    global TN1
    global FP1
    global FN1
    global TPR1
    global TNR1
    global FPR1
    global NPV1
    global FDR1
    global acc1
    global prec1
    global rec1
    y_pred=[]
    for x in X:
        positive=0
        negative=0

        for p in pluses:
            res=p & x
            positive+=float(len(res)) / len(p)
        positive=float(positive) / len(pluses) 

        for m in minuses:
            res=m & x
            negative+=float(len(res)) / len(m)
        negative=float(negative) / len(minuses) 

        if (positive > negative):
          y_pred.append(1)
        else:
          y_pred.append(0)

    y_pred = np.array(y_pred)

    TP = np.sum(y * y_pred)
    TN = np.sum(y + y_pred == 0)
    FP = np.sum((y  == 0) * (y_pred == 1))
    FN = np.sum((y  == 1) * (y_pred == 0))
    TPR = float(TP) / np.sum(y == 1)
    TNR = float(TN) / np.sum(y == 0)
    FPR = float(FP) / (TP + FN)
    NPV = float(TN) / (TN + FN)
    FDR = float(FP) / (TP + FP)
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred)
    rec = recall_score(y, y_pred)

    TP1 += TP
    TN1 += TN
    FP1 += FP
    FN1 += FN
    TPR1 += TPR
    TNR1 += TNR
    FPR1 += FPR
    NPV1 += NPV
    FDR1 += FDR
    acc1 += acc
    prec1 += prec
    rec1 += rec

    print ("True Positive: {}\nTrue Negative: {}\nFalse Positive: {}\n\
False Negative: {}\nTrue Positive Rate: {}\nTrue Negative Rate: {}\n\
Negative Predictive Value: {}\nFalse Positive Rate: {}\nFalse Discovery Rate: {}\nAccuracy:{}\nPrecision: {}\n\
Recall: {}".format(TP, TN, FP, FN, TPR, TNR, FPR, NPV, FDR, acc, prec, rec))
    
    return accuracy_score(np.array(y), np.array(y_pred))
```

```python
print('Algorithm0:')
avg = 0
for i in range(1,11):
    print ('Train set: ', i)
    plus, minus, x_test, y_test = read_data(i,i)
    pred=algorithm0(plus, minus, x_test, y_test)
    avg +=pred
    print('Test accuracy: ', round(pred,5))
    print()
print("Average TP: ", TP1/10)
print("Average TN: ", TN1/10)
print("Average FP: ", FP1/10)
print("Average FN: ", FN1/10)
print("Average TPR: ", TPR1/10)
print("Average TNR: ", TNR1/10)
print("Average FPR: ", FPR1/10)
print("Average NPV: ", NPV1/10)
print("Average FDR: ", FDR1/10)

print("Average accuracy: ", acc1/10)
print("Average recall: ", prec1/10)
print("Average recall: ", rec1/10)
```

```python
TP1 = 0
TN1 = 0
FP1 = 0
FN1 = 0
TPR1 = 0
TNR1 = 0
FPR1 = 0
NPV1 = 0
FDR1 = 0
acc1 = 0
prec1 = 0
rec1 = 0
def algo3(pluses, minuses, X, y, param):
    global TP1
    global TN1
    global FP1
    global FN1
    global TPR1
    global TNR1
    global FPR1
    global NPV1
    global FDR1
    global acc1
    global prec1
    global rec1
    y_pred=[]
    for t in X:
        
        positive=0
        negative=0

        for p in pluses:
            res=p & t
            if len(res)!=0:
                rate=0.0
                for k in pluses:
                    if k.issuperset(res) and k!=p:
                        rate+=1
                if rate>param:
                    positive+= float(rate) / len(pluses) * float(len(res)) / len(p)
                        
        positive=  float(positive) / len(pluses) 

        for m in minuses:
            res=m & t
            if len(res)!=0:
                rate=0
                for k in minuses :
                    if k.issuperset(res) and k!=m:
                        rate+=1
                if rate>param:        
                    negative+=float(rate) / len(minuses) * float(len(res)) / len(m) 
                
        negative =  float(negative) / len(minuses) 
        
           
        if (positive > negative):
            y_pred.append(1)
        else:
            if (positive< negative):
                y_pred.append(0)
            else:
                y_pred.append(0.5)

    y_pred = np.array(y_pred)

    TP = np.sum(y * y_pred)
    TN = np.sum(y + y_pred == 0)
    FP = np.sum((y  == 0) * (y_pred == 1))
    FN = np.sum((y  == 1) * (y_pred == 0))
    TPR = float(TP) / np.sum(y == 1)
    TNR = float(TN) / np.sum(y == 0)
    FPR = float(FP) / (TP + FN)
    NPV = float(TN) / (TN + FN)
    FDR = float(FP) / (TP + FP)
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred)
    rec = recall_score(y, y_pred)

    TP1 += TP
    TN1 += TN
    FP1 += FP
    FN1 += FN
    TPR1 += TPR
    TNR1 += TNR
    FPR1 += FPR
    NPV1 += NPV
    FDR1 += FDR
    acc1 += acc
    prec1 += prec
    rec1 += rec

  
    return accuracy_score(np.array(y), np.array(y_pred))
```

```python
print('Algorithm3:')
avg = 0;
for i in range(1,11):
    plus, minus, x_test, y_test = read_data(i,i)
    pred=algo3(plus, minus, x_test, y_test, 8)
    avg+=pred
```
